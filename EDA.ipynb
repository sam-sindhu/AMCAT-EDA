```python
# Detect outliers using IQR or other methods
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

# Filter out outliers based on IQR (optional)
df = df[(df['Salary'] >= Q1 - 1.5 * IQR) & (df['Salary'] <= Q3 + 1.5 * IQR)]

# Verify the number of rows after removing outliers
print(f"\nNumber of rows after removing outliers: {df.shape[0]}")
```

```python
# 1. Histograms and Boxplots for Numerical Columns
import seaborn as sns

# Automatically select all numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Function to detect outliers in a column using IQR method
def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    return outliers

# Loop over each numerical column to find and print outliers
for col in numerical_cols:
    outliers = detect_outliers(df, col)
    print(f"Number of outliers in {col}: {outliers.shape[0]}")
    if not outliers.empty:
        print(f"Examples of outliers in {col}:")
        print(outliers[[col]].head(), "\n")
```

```python
#Step - 5 - Research Questions
#Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering 
#if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer 
#you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.

cs_graduates = df[df['Specialization'] == 'computer science & engineering']
roles = ['programming analyst', 'software engineer', 'hardware engineer', 'associate engineer']

cs_graduates_roles = cs_graduates[cs_graduates['Designation'].isin(roles)]

print(cs_graduates_roles[['Designation', 'Salary']].describe())

within_range = cs_graduates_roles[(cs_graduates_roles['Salary'] >= 2.5) & (cs_graduates_roles['Salary'] <= 3.0)]

print(f"Number of freshers in these roles earning between 2.5 and 3 lakhs: {within_range.shape[0]}")
print(f"Total number of freshers in these roles: {cs_graduates_roles.shape[0]}")
print(f"Percentage of freshers earning 2.5-3 lakhs: {(within_range.shape[0] / cs_graduates_roles.shape[0]) * 100:.2f}%")
```

```python
#Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)
gender_specialization_crosstab = pd.crosstab(df['Gender'], df['Specialization'])

# Perform Chi-Square Test of Independence
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(gender_specialization_crosstab)

# Step 3: Output the results
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")

# Interpretation
if p < 0.05:
    print("There is a significant relationship between Gender and Specialization (reject the null hypothesis).")
else:
    print("There is no significant relationship between Gender and Specialization (fail to reject the null hypothesis).")
```
